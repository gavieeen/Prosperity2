```python
from typing import Tuple, List
from math import log
import numpy as np

# Given rates as a list
rates2 = [
    [[100, 48, 152, 71], [205, 100, 326, 156], [64, 30, 100, 46], [141, 61, 208, 100]]
]

rates = np.array(rates2)
rates = rates / 100.0

max_profit = 1.0

# Starting index
start_index = 3

# Iterate over possible trade destinations
for j in range(4):  # First trade destination
    curr_profit = 1.0 * rates[0][start_index][j] * rates[0][j][start_index]

    if curr_profit > max_profit:
        print("New Combo:")
        print(j)
        max_profit = curr_profit
    for k in range(4):  # Second trade destination
        curr_profit = (
            1.0 * rates[0][start_index][j] * rates[0][j][k] * rates[0][k][start_index]
        )
        
        if curr_profit > max_profit:
            print("New Combo:")
            print(j)
            print(k)
            max_profit = curr_profit
        for l in range(4):  # Third trade destination
            curr_profit = (
                    1.0
                    * rates[0][start_index][j]
                    * rates[0][j][k]
                    * rates[0][k][l]
                    * rates[0][l][start_index]
            )
            
            if curr_profit > max_profit:
                print("New Combo:")
                print(j)
                print(k)
                print(l)
                max_profit = curr_profit
            for m in range(4):  # Fourth trade destination
                # Check if trades end at index 3
                curr_profit = (
                    1.0
                    * rates[0][start_index][j]
                    * rates[0][j][k]
                    * rates[0][k][l]
                    * rates[0][l][m]
                    * rates[0][m][start_index]
                )

                if curr_profit > max_profit:
                    print("New Combo: ")
                    print(j)
                    print(k)
                    print(l)
                    print(m)
                    max_profit = curr_profit

# Output the maximum profit
print("Max Profit:", max_profit)
```
